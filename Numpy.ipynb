### <span style = "font-size:30px"> **(3) : What is the purpose of the axis parameter in functions like `numpy.sum()` ?**

```python
# it lets you do the function through the whole array, or just dealing with columns or rows

num_array = np.array( [[1, 2], [3, 4]])
# example - sum of all elements in the array
total_sum = np.sum(num_array)
print("Total Sum:", total_sum)

# example - sum of each column
column_sum = np.sum(num_array, axis=0)
print("Column Sum:", column_sum)

# example - sum of each row
row_sum = np.sum(num_array, axis=1)
print("Row Sum:", row_sum)
```

### <span style = "font-size: 30px"> **(4) : Describe the difference between using `*` (multiplication operator) and `@` (matrix multiplication operator) when operating on two `2D NumPy arrays`.**

```python

#* THe sign * is used to perform corresponding element multiplication between two arrays.
test_array1 = np.array([4, 7, 3])
test_array2 = np.array([2, 5, 8])
result_array = test_array1 * test_array2 #? will be multibly each element with the corresponding element and store them in a new array
print(result_array)

#* The sign @ performs matrix multiplication - #!    
#* each row of the first matrix is multiplied with each column of the second matrix and the results are summed to produce a new matrix
test_matrix1 = np.array([[1, 2], [3, 4]])
test_matrix2 = np.array([[5, 6], [7, 8]])
result_array = test_matrix1 @ test_matrix2
print(result_array)
```
